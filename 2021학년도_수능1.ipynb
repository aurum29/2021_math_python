```python
from itertools import permutations as p
from fractions import Fraction as F

X = ['A', 'B', 'C', 'D', 'E', 1, 2, 3, 4]
total = list(p(X))  # 전체 나열

print(len(total))   # 9!

Y = []
for i in total:
    a = i.index('A')
    if a != 0 and a != 8:
        if type(i[a-1]) == int and type(i[a+1]) == int:
            Y.append(i)

print(len(Y))

print(F(len(Y),len(total)))
```
