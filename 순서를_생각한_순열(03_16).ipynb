# 순서를 생각한 순열

- 같은 것을 포함한 순열

    a a a b b: 5!/(3! 2!)
    - 문자가 두개 일때 조합! : 5C3
    - but, 문자가 3개 이상일 때 조합x



```
# '지우' < '성민' < '지형' 순서로 먼저 배치하고 나머지 배치하기(순열)

from itertools import permutations as p     # itertools 이라는 모듈에서 permutations 가져옴

A = ['의성', '지우', '성민', '남호', '제욱', '지형']    
B = list(p(A))
C = []
for i in B:
    if i.index('지형') > i.index('성민') > i.index('지우'):
        C.append(i)

print(C)
print(len(C))
D = [', '.join(i) for i in C]
print(D)
```
