```python
from fractions import Fraction as F

nu = [3,3,4,4,4]
print(nu.count(3), nu.count(4))  # 3의 갯수

print(F(nu.count(3),len(nu)))
print(F(nu.count(4),len(nu)))
```

```python
from itertools import product as pr

dice3 = list(pr(range(1, 7), repeat=3))
d3 = [] # 주사위 3번 던져 합이 10
# print(dice3)
# print(len(dice3))

dice4 = list(pr(range(1, 7), repeat=4))
d4 = [] # 주사위 4번 던져 합이 10
# print(dice4)
# print(len(dice4))

d3 = [ i for i in dice3 if sum(i) == 10]
d4 = [ i for i in dice4 if sum(i) == 10]

df3 = F(len(d3), len(dice3))
df4 = F(len(d4), len(dice4))

# print(df3, df4)

print(F(nu.count(3),len(nu))*df3+F(nu.count(4),len(nu))*df4)
# 3이 나올 확률 * 3번 던져서 합이 10될 확률 + 4가 나올 확률 * 4번 던져서 합이 10이 될 확률
```
