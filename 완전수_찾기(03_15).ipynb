# 완전수 찾기

완전수 : 진약수의 합과 비교

```
def divisor(x):     # 진약수의 합을 계산하는 함수 정의
    a = 0
    for i in range(1, x):   # 1<= i <x
        if x % i == 0:
            a += i
    return a            
```

```
MAX = int(input())
A = []
for i in range(1,MAX+1):
    if divisor(i) == i:
        print(i)
        A.append(i)
print(A)
```

```
B = list(i for i in range(1,1000) if divisor(i)==i)
print(B)
```
