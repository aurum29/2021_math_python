```python
from itertools import combinations_with_replacement as H
from itertools import product as pr

BB = ['Ab', 'Bb', 'Cb', 'Db']   # Ab : A가 black를 갖는 경우
WW = ['Aw', 'Bw', 'Cw', 'Dw']   # Aw : A가 white를 갖는 경우
totalB = list(H(BB,6))
totalW = list(H(WW,6))
totalBW = list(pr(totalB, totalW))

#print('4H4 = ', len(totalB))    # 84
#print(totalBW)  # totalBW[0] = (('Ab', 'Ab', 'Ab', 'Ab', 'Ab', 'Ab'), ('Aw', 'Aw', 'Aw', 'Aw', 'Aw', 'Aw'))

BW = [i[0] + i[1] for i in totalBW]

# print(BW)   # BW[0] = ('Ab', 'Ab', 'Ab', 'Ab', 'Ab', 'Ab', 'Aw', 'Aw', 'Aw', 'Aw', 'Aw', 'Aw')

WB4 = []
for i in BW:
    if set([j[0] for j in i]) == {'A', 'B', 'C', 'D'} and i.count('Ab') >= 4 and i.count('Ab') > i.count('Aw'):   # (가) and (나)
        WB4.append(i)
    '''
    k = []
    for j in i:
        k.append(j[0])  # string 원소의 첫번째 자리를 의미
    
    if len(set(k)) == 4:
        WB4.append(i)
    '''

#print(len(WB4))

WBtotal = []
for i in WB4:
    if (i.count('Bb') > i.count('Bw')) + (i.count('Cb') > i.count('Cw')) + (i.count('Db') > i.count('Dw')) == 1: # 하나만 True 인 경우
        WBtotal.append(i)

print(len(WBtotal))
```
