```python
from itertools import combinations as c
from itertools import permutations as p
from fractions import Fraction as F
X = [1, 2, 3, 4]
powerX = [] # X의 멱집합
for i in range(1,5):
    powerX.extend(c(X, i))  # append 와 잘 구분하자

print(powerX)
print('부분집합의 갯수 :', len(powerX))

superX = list(p(powerX, 3))
print(superX)
print('전체 경우의 수 :', len(superX))  # 15*14*13

# 성주 풀이
lastX = [i for i in superX if set(i[0]).issubset(set(i[1])) and set(i[1]).issubset(set(i[2]))]
print(F(len(lastX),len(superX)))

# 쌤 풀이
lastX = []
for i in superX:
    if all(j in i[1] for j in i[0]) and all(j in i[2] for j in i[1]):
        lastX.append(i)
        #print(i)
print(F(len(lastX), len(superX)))
```

```python
# practice
a = [1, 2, 3,7]
b = [1, 2, 3, 4, 5]
for i in a:
    if i in b:
        print(i, 'in b')
    else:
        print(i, 'not in b')

print('\n', any(i in b for i in a)) # any는 or을 의미
print(all(i in b for i in a))   # all은 and를 의미
```
