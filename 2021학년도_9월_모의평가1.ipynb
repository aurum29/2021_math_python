```python
from itertools import permutations as p
import itertools
total = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
total1 = list(p(total[2:],3))
totalAB = list(p(total[0:2]))

final = list(itertools.product(total1, totalAB))
# final = [('A', 'B') + i for i in total1] + [('B', 'A') + i for i in total1]

print(len(final))
```

```python
# 원순열(추가 설명)
total = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']    # H를 고정
ex = list(p(total[:7]))
ex2 = [i + ('H',) for i in ex]
print(ex2)
```

```python
# 더 간단한 방법
from itertools import permutations as p
Student=['A','B','C','D','E','F','G','H']
All_case=list(p(Student, 5))
Case = [i for i in All_case if i[0] == 'A' and (i[1] == 'B' or i[-1] == 'B')]   # 순열을 구한 뒤에 중복되는 경우 제외
print(len(Case))
```
