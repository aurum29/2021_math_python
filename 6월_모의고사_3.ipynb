 두 집합 A = {1, 2, 3, 4}, B = {1, 2, 3} 에 대하여 A 에서 
B로의 모든 함수 f 중에서 임의로 하나를 선택할 때, 
이 함수가 다음 조건을 만족시킬 확률은?

f(1) ≥ 2 이거나 함수 f 의 치역은 B이다.

```python
from itertools import product as pr
from fractions import Fraction as F
A = [1, 2, 3]

B = list(pr(A, repeat = 4))
# print(B)
# print(len(B))   # 3**4

C = []
for i in B:
    a = set(i)
    if i[0] >= 2 or A == list(a):   # A == list(a) 는 len(a) == 3 와 같다.
        C.append(i)
print(C)
print(F(len(C),len(B)))
```
